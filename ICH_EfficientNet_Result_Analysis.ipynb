```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_results.csv')
```

```python
adf = pd.read_csv('analysis_final_results.csv')
```

```python
print(df.shape)
df.head(5)
```

```python
print(adf.shape)
adf.head(5)
```

```python
df['img_file'] = df['ID'].apply(lambda st: "ID_" + st.split('_')[1] + ".png")
df['subtype'] = df['ID'].apply(lambda st: st.split('_')[2])
groupSub = df.groupby('subtype').sum()
label_subtype_plot = sns.barplot(y=groupSub.Label, x=groupSub.index, palette="Blues_d")
```

```python
#Analysis on the hemorrhage subtype
a_groupSub = adf.groupby('Hemorrhage Subtype').sum()
a_label_subtype_plot = sns.barplot(y=a_groupSub.Probability, x=a_groupSub.index, palette="muted")
```

```python
#pie chart analysis of final_results.csv
fig1, ax1 = plt.subplots()
ax1.pie(groupSub.Label, labels=groupSub.index, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
#pie chart analysis of analysis_final_results.csv

fig1, ax1 = plt.subplots()
ax1.pie(a_groupSub.Probability, labels=a_groupSub.index, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
